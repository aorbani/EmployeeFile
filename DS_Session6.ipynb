Pandas is a powerful, easy, and an open source python library. The usage of this tool is to structure table data and manupilate it. Which makes it most widely used for data science/data analysis and machine learning tasks

```python
import pandas as pd
import numpy as np
```

Creating Series from array

```python

readings = np.array([841,514,126,975,326,121,422])
print(readings)

readings_sr = pd.Series(readings)
print(readings_sr)

actual_readings = pd.Series([1, 3, 5, np.nan, 6, 8])
print(readings_sr)
```

Creating Series from dictionary

```python
calories = {"day1": 420, "day2": 380, "day3": 390}
calories_sr = pd.Series(calories)
print(calories_sr)
```
